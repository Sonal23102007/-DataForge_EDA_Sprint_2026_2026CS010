```python
import os

os.makedirs("04_python/outputs/tables", exist_ok=True)
```

```python
import pandas as pd

df = pd.read_csv("01_data/processed/dataforge_cleaned.csv")

df.head()
```

```python
numeric_summary = df.describe().T

numeric_summary.to_csv("04_python/outputs/tables/numeric_summary.csv")

numeric_summary
```

```python
missing = (df.isna().mean() * 100).sort_values(ascending=False)

missing.to_csv("04_python/outputs/tables/missing_percent.csv")

missing
```

```python
sales_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

sales_region.to_csv("04_python/outputs/tables/sales_by_region.csv")

sales_region
```

```python
sales_segment = df.groupby("Segment")["Sales"].sum().sort_values(ascending=False)

sales_segment.to_csv("04_python/outputs/tables/sales_by_segment.csv")

sales_segment
```

```python
sales_year = df.groupby("year")["Sales"].sum().sort_values(ascending=False)

sales_year.to_csv("04_python/outputs/tables/sales_by_year.csv")

sales_year
```

```python
top10 = df.sort_values(by="Sales", ascending=False).head(10)

top10.to_csv("04_python/outputs/tables/top10_sales.csv")

top10
```

```python
bottom10 = df.sort_values(by="Sales", ascending=True).head(10)

bottom10.to_csv("04_python/outputs/tables/bottom10_sales.csv")

bottom10
```

1. Region with highest sales contributes the majority of total revenue.
2. Certain categories generate higher profit compared to others.
3. Sales distribution varies significantly across segments.
4. Yearly sales trend shows variation indicating possible growth or decline.
5. Top 10 sales transactions are significantly higher than bottom 10, showing skewness.
